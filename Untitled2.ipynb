```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/Carbon Emission.csv')
print(df)
```

```python
df.info()
```

```python
# Drop missing values
df_subset = df[['Diet', 'Transport', 'Heating Energy Source', 'CarbonEmission']].dropna()
print(df_subset)
```

```python
# One-hot encode categorical inputs (fixed for scikit-learn ≥ 1.2)
encoder = OneHotEncoder(sparse_output=False)
X_encoded = encoder.fit_transform(df_subset[['Diet', 'Transport', 'Heating Energy Source']])
encoded_feature_names = encoder.get_feature_names_out(['Diet', 'Transport', 'Heating Energy Source'])
```

```python
# Feature matrix
X = pd.DataFrame(X_encoded, columns=encoded_feature_names)
y = df_subset['CarbonEmission']  # CO2 in kg/month

# Convert to kg/week
y_week = y / 4.3

# Split dataset
X_train, X_test, y_train_week, y_test_week = train_test_split(X, y_week, test_size=0.2, random_state=42)

# Train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train_week)

# Predict and evaluate
y_pred_week = model.predict(X_test)
mse = mean_squared_error(y_test_week, y_pred_week)
print(f"\nMean Squared Error (MSE): {mse:.2f} kg²/week²")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_test_week, y=y_pred_week, color='green', s=100)
plt.plot([y_test_week.min(), y_test_week.max()],
         [y_test_week.min(), y_test_week.max()], 'r--', linewidth=2)
plt.xlabel('Actual CO₂ Emissions (kg/week)')
plt.ylabel('Predicted CO₂ Emissions (kg/week)')
plt.title('Actual vs Predicted Weekly Carbon Emissions')
plt.grid(True)
plt.tight_layout()
plt.show()
```
